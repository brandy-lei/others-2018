### get second tier url link in article

```python
newurl=[]
for anchor in soup.find_all("a"):
        # extract link url from the anchor
    link = anchor.attrs["href"] if "href" in anchor.attrs else ''
    newurl.append(link)
newurl
```

```python
# http://blog.csdn.net/chichoxian/article/details/72362969
np.random.seed(19680801)

mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000)

# the histogram of the data
n, bins, patches = plt.hist(x, 50, normed=1, facecolor='g', alpha=0.75)
plt.text(60, .025, '$\mu=100, \sigma=15$')
plt.axis([40,160,0,0.03])
plt.grid(True)
plt.xlabel('my data', fontsize=14, color='red')
```

```python
# http://blog.csdn.net/chichoxian/article/details/72362969
#fig = plt.figure(figsize=(6,4), dpi=100)
np.random.seed(19680801)

mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000)
bins = np.linspace(min(x),max(x),50)
# the histogram of the data
n, bins, patches = plt.hist(x, bins, normed=1, facecolor='g', alpha=0.75)
plt.text(60, .025, '$\mu=100, \sigma=15$')
plt.axis([40,160,0,0.03])
plt.grid(True)
plt.xlabel('my data', fontsize=14, color='red')
```
